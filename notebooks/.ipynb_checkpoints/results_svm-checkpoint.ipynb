```python
import numpy as np
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

SVM
C = 1000
gamma = 0.1

```python
x_plot = range(0, 1000, 9)

f = open('svm_results.p', 'rb')
y_plot = pickle.load(f)

f = open('svm_errors.p', 'rb')
error = pickle.load(f)


fig, ax = plt.subplots()
ax.plot(x_plot, y_plot)

plt.yticks(np.arange(0.0, 1.05, 0.05))

x_labels = ax.get_xticks()
ax.set_xticklabels(['{:1.1f}'.format(float(x)/1000) for x in x_labels])


plt.grid(linestyle='dotted')
plt.ylabel('Prediction accuracy')
plt.xlabel('Relative time')


plt.show()
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_plot, y_plot, yerr=error, fmt='', capsize=2)

plt.yticks(np.arange(0.0, 1.05, 0.05))

x_labels = ax.get_xticks()
ax.set_xticklabels(['{:1.1f}'.format(float(x)/1000) for x in x_labels])


plt.grid(linestyle='dotted')
plt.ylabel('Prediction accuracy')
plt.xlabel('Relative time')


plt.show()
```
